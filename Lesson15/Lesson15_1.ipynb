```python
student1 = {'chinese':78, 'english':80, 'math':92, 'history':75, 'discover':85}
student2 = {'chinese':92, 'english':60, 'math':85, 'history':62, 'discover':73}
keys = student1.keys()
subject:list[str] = list(keys)
values = student1.values()
values1 = list(values)
values2 = list(student2.values())

#
import matplotlib.pyplot as plt

figure = plt.figure(figsize=(8,5),dpi=250,facecolor='#EEA8A8')
axes = figure.add_subplot(1,1,1)
#
axes.plot(subject,values1,'co--',label='student1')   #
axes.plot(subject,values2,'yo-.',label='student2')   #
axes.grid(axis='y')
axes.set_title("student1 vs student2")  #
axes.set_xlabel('subjects')     #X
axes.set_ylabel('score')        #Y
axes.legend()
#print(axes.get_xticks())    #X
#print(axes.get_yticks())    #Y
labels = axes.get_xticklabels()
for label in labels:
    subject_name = label.get_text()     #X
    score1 = student1[subject_name]
    x_value = label.get_position()[0]  #X
    axes.annotate(str(score1),(x_value,score1+0.5))
    score2 = student2[subject_name]
    axes.annotate(str(score2),(x_value,score2+0.5))
plt.show()
```
